```python
IP = "127.0.0.1" #localhost, mi propia computadora
PUERTO = 9001

client = SimpleUDPClient(IP, PUERTO) 
```

```python
df1 = df[ ["Perro", "Mujer", "Varon", "Barrio"] ]
df2 = df[ ["Perro", "Mujer", "Varon", "Barrio"] ]
df3 = df[ ["Perro", "Mujer", "Varon", "Barrio"] ]

dfConcat = pd.DataFrame( np.concatenate( [df1.values, df2.values, df3.values], axis=0 ), 
                                          columns=["Perro","Mujer","Varon","Barrio"] )

dfConcat
```

```python
listaAnimal = dfConcat["Perro"]
listaHumanes = (dfConcat["Mujer"] + dfConcat["Varon"])


listaHumanes
```

```python
client.send_message("/animales", listaAnimal[6])
```

```python
def tempo(t1, t2, l1, l2):
  def tempo1(t1, l1):
    for i in range(len(l1)):
        return time.sleep(t1), client.send_message("/humanes", l1[i])
        
  def tempo2(t2, l2):
    for i in range(len(l2)):
        return time.sleep(t2), client.send_message("/animales", l2[i])
        
  
  return tempo1(t1, l1), tempo2(t2, l2)
```

```python
while True:
    tempo(0.5, 4, listaHumanes, listaAnimal )
```

```python
for i in range(len(listaHumanes)):
    time.sleep(1.5)
    client.send_message("/humanes", listaHumanes[i])
    time.sleep(4.5)
    client.send_message("/animales", listaAnimal[i])
```
